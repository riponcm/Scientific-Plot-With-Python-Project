```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
# Load the CSV data
data = pd.read_csv(r'C:\Users\Ripon\Desktop\python\data.csv')
```

```python
# Function to create a smooth line
def smooth_line(x, y):
    x_new = np.linspace(x.min(), x.max(), 300)  # Adjust number of points as needed
    f = interpolate.interp1d(x, y, kind='cubic')
    y_smooth = f(x_new)
    return x_new, y_smooth
```

```python
# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 8), sharex=True, sharey=True)

# Plot the data with smooth lines
#subplot_1
x1, y1_a1 = smooth_line(data['x1'], data['a1'])
x1, y1_a2 = smooth_line(data['x1'], data['a2'])
axs[0, 0].plot(x1, y1_a1, label='HVDM Treated', color='blue')
axs[0, 0].plot(x1, y1_a2, label='Remoulded', color='red')
axs[0, 0].set_title('Sample 1')

#subplot_2
x2, y2_b1 = smooth_line(data['x2'], data['b1'])
x2, y2_b2 = smooth_line(data['x2'], data['b2'])
axs[0, 1].plot(x2, y2_b1, label='a1', color='blue')
axs[0, 1].plot(x2, y2_b2, label='a2', color='red')
axs[0, 1].set_title('Sample 2')


#subplot_3
x3, y3_c1 = smooth_line(data['x3'], data['c1'])
x3, y3_c2 = smooth_line(data['x3'], data['c2'])
axs[1, 0].plot(x3, y3_c1, label='a1', color='blue')
axs[1, 0].plot(x3, y3_c2, label='a2', color='red')
axs[1, 0].set_title('Sample 3')
#subplot_4
x4, y4_d1 = smooth_line(data['x4'], data['d1'])
x4, y4_d2 = smooth_line(data['x4'], data['d2'])
axs[1, 1].plot(x4, y4_d1, label='a1', color='blue')
axs[1, 1].plot(x4, y4_d2, label='a2', color='red')
axs[1, 1].set_title('Sample 4')

# Set common axis labels
fig.text(0.5, -0.02, 'Strain (%)', ha='center')
fig.text(-0.02, 0.5, 'Stress (kPa)', va='center', rotation='vertical')

# Adjust layout
plt.tight_layout()

# Add legend to the first subplot
axs[0, 0].legend(loc='best')

#Save the figure to the Desktop 
import os 
desktop_path = os.path.join(os.path.expanduser('~'), 'Desktop')
save_path = os.path.join(desktop_path, 'stressvsstrain.png')
plt.savefig(save_path, dpi=300)
plt.show()
```
